# 4. 시각화(EDA)

```python
df_mart['월'] = df_mart['공급일자_dt'].dt.month  # 월 추출하여 새로운 열 추가
df_mart['주'] = df_mart['공급일자_dt'].dt.isocalendar().week  # 주 추출하여 새로운 열 추가
df_mart['일'] = df_mart['공급일자_dt'].dt.day  # 일 추출하여 새로운 열 추가
df_mart['요일'] = df_mart['공급일자_dt'].dt.day_name()  # 요일 추출하여 새로운 열 추가
```

```python
df_mart.head(1)
```

```python
# 월별 구입금액을 확인
pd.pivot_table(data=df_mart, index='월', values='구매금액', aggfunc='sum')
```

```python
# 한글 폰트 설정
import matplotlib.pyplot as plt
plt.rc('font', family='Malgun Gothic')  # Windows
plt.rcParams['axes.unicode_minus'] = False  # 음수 기호 표시
```

```python
 # 막대 그래프로 시각화
pd.pivot_table(data=df_mart, index='월', values='구매금액', aggfunc='sum').plot(kind='bar', figsize=(10, 5), title='월별 구입금액') 
```

```python
# 주별 구입금액을 확인
pd.pivot_table(data=df_mart, index='주', values='구매금액', aggfunc='sum').plot(kind='bar', figsize=(10, 5), title='주별 구입금액')
```

```python
# 요일별 구입금액을 확인
pd.pivot_table(data=df_mart, index='요일', values='구매금액', aggfunc='sum').plot(kind='bar', figsize=(10, 5), title='요일별 구입금액')
```

```python
plt.figure(figsize=(10, 5))
sns.countplot(data=df_mart, x='물품대분류', hue='연령대')
plt.legend(loc='right')
plt.savefig('./age2_chart.pdf')
plt.savefig('./age2_chart.png')
```
